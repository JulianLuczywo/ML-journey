## Lesson 3 - Bar Charts and Heatmaps

```python
flight_filepath = "../Data-Visualisation/flight_delays.csv"

flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
flight_data
```

### Bar Charts

```python
plt.figure(figsize=(10,6))

plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

sns.barplot(x=flight_data.index, y=flight_data['NK'])

plt.ylabel("Arrival delay (in minutes)")

plt.savefig('figures/05.png', dpi=300)

plt.show()
```

### Heatmap

```python
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for Each Airline, by Month")

# annot=True -> ensures that the values for each cell appear on the chart.
sns.heatmap(data=flight_data, annot=True)

plt.xlabel("Airline")

plt.savefig('figures/06.png', dpi=300)

plt.show()
```

## Lesson 4 - Scatter Plots

```python
insurance_filepath = "../Data-Visualisation/insurance.csv"

insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])

plt.savefig('figures/07.png', dpi=300)

plt.show()
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])

plt.savefig('figures/08.png', dpi=300)

plt.show()
```

### Colour-scoded Scatter Plots

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])

plt.savefig('figures/09.png', dpi=300)

plt.show()
```

`sns.lmplot` adds two regresssion lines, corresponding to smokers and nonsmokers

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)

plt.savefig('figures/10.png', dpi=300)

plt.show()
```

### Categorical Scatter Plot

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])

plt.savefig('figures/11.png', dpi=300)

plt.show()
```

Plot shows us that:
- on average, non-smokers are charges less than smokers
- customers who pay the most are smokers; wheareas the customers who pay the least are non-smokers

## Lesson 4 - Distributions

```python
iris_filepath = "../Data-Visualisation/iris.csv"

iris_data = pd.read_csv(iris_filepath, index_col="Id")

iris_data.head()
```

### Histogram

```python
sns.histplot(iris_data['Petal Length (cm)'])

plt.savefig('figures/12.png', dpi=300)

plt.show()
```

### Density Plot

**Kernel Density Estimate (KDE)** - smoothed histogram `sns.kdeplot`

```python
# KDE plot

sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True)

plt.savefig('figures/13.png', dpi=300)

plt.show()
```

###2D KDE Plots

**Two-dimensional (2D) KDE plot** - `sns.joinplot`

```python
# 2D KDE plot

sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")

plt.savefig('figures/14.png', dpi=300)

plt.show()
```

- Curve at the top of the figure is a KDE plot for the data on the x-axis (`iris_data['Petal Length (cm)']`)
- Curve on the right of the figure is a KDE plot of the data on the y-axis (`iris_data['Sepal Width (cm)']`)

## Colour-coded Plots

- `data` provides the name of the vairables that we used to read in the data
- `x` sets the name of column with the data we want to plot
- `hue` sets the column we'll use to split the data into different histograms

```python
# Histograms for each species

sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

plt.title("Histogram of Petal Lengths, by Species")

plt.savefig('figures/15.png', dpi=300)

plt.show()
```

```python
# KDE plots for each species

sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)

plt.title("Distribution of Petal Lengths, by Species")

plt.savefig('figures/16.png', dpi=300)

plt.show()
```
