```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting Functions and Lines

```python
x = np.linspace(0, 10, 100) # 100 values between 0-10
plt.plot(x, np.sin(x))

plt.show()
```

Added axis ranges and labels:

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))

plt.xlim([2, 8]) # limits for x-axis
plt.ylim([0, 0.75]) # limits for y-axis

plt.xlabel('x-axis') # x-axis label
plt.ylabel('y-axis') #Â y-axis label

plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label=('sin(x)'), color='blue')
plt.plot(x, np.cos(x), label=('cos(x)'), color='red')

plt.ylabel('f(x)')
plt.xlabel('x')

plt.legend(loc='lower left')
plt.show()
```

### Scatter Plot

```python
rng = np.random.RandomState(123)
x = rng.normal(size=500)
y = rng.normal(size=500)

plt.scatter(x, y)
plt.show()
```

### Bar Plots

```python
# input data
means = [1, 2, 3]
stddevs = [0.2, 0.4, 0.5]
bar_labels = ['bar 1', 'bar 2', 'bar 3']

# plot bars
x_pos = list(range(len(bar_labels)))
plt.bar(x_pos, means, yerr=stddevs)

plt.show()
```

### Histograms

```python
# input data
rng = np.random.RandomState(123)
x = rng.normal(0, 20, 1000)

# fixed bin size
bins = np.arange(-100, 100, 5)

# plot histogram
plt.hist(x, bins=bins)

plt.show()
```

```python
# input data
rng = np.random.RandomState(123)
x1 = rng.normal(0, 20, 1000)
x2 = rng.normal(15, 10, 1000)

# fixed bin size
bins = np.arange(-100, 100, 5)

# plot histogram
plt.hist(x1, bins=bins, alpha=0.5)
plt.hist(x2, bins=bins, alpha=0.5)

plt.show()
```

### Colours and Markers

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x),
         color='green',
         marker='*',
         linestyle='')

plt.show()
```

### Saving Plots

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))

plt.savefig('images/myplot.png',dpi=300)
plt.savefig('images/myplot.pdf')

plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

line_style = dict(marker=".",
               markersize=30,
               markerfacecolor="#899878",
               markeredgecolor="#899878",
               linestyle="solid",
               linewidth=3)

plt.plot(x, y1, color="#222725", **line_style)
plt.plot(x, y2, color="#1cfc45", **line_style)
plt.plot(x, y3, color="#fc1c1c", **line_style)

plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

label_style = dict(fontsize=20, fontweight="bold")

plt.title("Class size", fontsize=25,
                        fontweight="bold",
                        color="#2d4cfc")

plt.xlabel('Year', **label_style)
plt.ylabel('Students', **label_style)

plt.tick_params(axis="both", color="red")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)

plt.show()
```
